# Download Mapillary

aria2c downloads multiple zips in parallel

```python
import subprocess
from pathlib import Path

Path("dataset").mkdir(exist_ok=True)

subprocess.run([
    "aria2c",
    "--input-file=links.txt",
    "--dir=dataset",
    "--max-concurrent-downloads=5",
    "--split=8",
    "--max-connection-per-server=8",
    "--min-split-size=20M",
    "--continue=true",
    "--console-log-level=notice",
], check=True)
```

```python
import zipfile
from pathlib import Path
from tqdm.notebook import tqdm

zip_files = sorted(Path("dataset").glob("*.zip"))
print(f"Found {len(zip_files)} zip files")

for zip_path in tqdm(zip_files):
    print(f"  Extracting {zip_path.name}...")
    with zipfile.ZipFile(zip_path, "r") as zf:
        zf.extractall("dataset")

print("Done.")
```

# Download LISA via roboflow

```python
import subprocess
from pathlib import Path
import zipfile

ROBOFLOW_URL = "https://universe.roboflow.com/ds/kK1J4x4m00?key=6bUhxEGINM"
DOWNLOAD_DIR = Path("dataset/roboflow_download")
DOWNLOAD_DIR.mkdir(parents=True, exist_ok=True)

# Roboflow dataset links download as a zip
zip_path = DOWNLOAD_DIR / "roboflow_dataset.zip"

subprocess.run([
    "aria2c",
    ROBOFLOW_URL,
    f"--dir={DOWNLOAD_DIR}",
    "--out=roboflow_dataset.zip",
    "--split=8",
    "--max-connection-per-server=8",
    "--continue=true",
    "--console-log-level=notice",
], check=True)

# Unzip into dataset/
print(f"Extracting {zip_path}...")
with zipfile.ZipFile(zip_path, "r") as zf:
    zf.extractall(DOWNLOAD_DIR)

print(f"Done. Contents:")
for p in sorted(DOWNLOAD_DIR.iterdir()):
    print(f"  {p.name}")
```
